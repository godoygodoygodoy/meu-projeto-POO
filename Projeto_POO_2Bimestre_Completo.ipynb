## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**

- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2

class Lanche_Godoy:
    def __init__(self, nome_lanche, preco, ingredientes):
        self.nome_lanche = nome_lanche
        self.preco = preco
        self.ingredientes = ingredientes

    def exibir_dados_Godoy(self):
        print(f"üçî Lanche: {self.nome_lanche}")
        print(f"üí∞ Pre√ßo: R${self.preco:.2f}")
        print("üìã Ingredientes:")
        for ingrediente in self.ingredientes:
            print(f" - {ingrediente}")
        print("-" * 30)

    def aplicar_desconto_Godoy(self, porcentagem):
        desconto = self.preco * (porcentagem / 100)
        self.preco -= desconto
        print(f"üîª Desconto de {porcentagem}% aplicado! Novo pre√ßo: R${self.preco:.2f}")

    def adicionar_ingrediente_Godoy(self, novo_ingrediente):
        self.ingredientes.append(novo_ingrediente)
        print(f"‚ûï Ingrediente '{novo_ingrediente}' adicionado com sucesso!")



class LancheCombo_Godoy(Lanche_Godoy):
    def __init__(self, nome_lanche, preco, ingredientes, bebida):
        super().__init__(nome_lanche, preco, ingredientes)
        self.bebida = bebida  # novo atributo

    def exibir_combo_Godoy(self):
        self.exibir_dados_Godoy()
        print(f"ü•§ Bebida inclusa: {self.bebida}")
        print("=" * 30)



batman_combo_Godoy = LancheCombo_Godoy(
    "Batman Combo",
    25.90,
    ["P√£o preto", "Hamb√∫rguer artesanal", "Queijo gruy√®re", "Molho secreto da caverna"],
    "Refrigerante de uva"
)


batman_combo_Godoy.exibir_combo_Godoy()
batman_combo_Godoy.aplicar_desconto_Godoy(10)
batman_combo_Godoy.adicionar_ingrediente_Godoy("Bat-bacon")
batman_combo_Godoy.exibir_combo_Godoy()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 3

class Lanche_Godoy:
    def __init__(self, nome_lanche, preco, ingredientes):
        self.nome_lanche = nome_lanche
        self.__preco = preco  # atributo privado
        self.ingredientes = ingredientes

    def exibir_dados_Godoy(self):
        print(f"üçî Lanche: {self.nome_lanche}")
        print(f"üí∞ Pre√ßo: R${self.__preco:.2f}")
        print("üìã Ingredientes:")
        for ingrediente in self.ingredientes:
            print(f" - {ingrediente}")
        print("-" * 30)

  
    def get_preco_Godoy(self):
        return self.__preco

    def set_preco_Godoy(self, novo_preco):
        if novo_preco > 0:
            self.__preco = novo_preco
            print(f"‚úÖ Pre√ßo atualizado para R${self.__preco:.2f}")
        else:
            print("‚ùå Pre√ßo inv√°lido! Tem que ser maior que zero.")

  
    def calcular_preco_final_Godoy(self, taxa=2.50):
        preco_final = self.__preco + taxa
        print(f"üì¶ Pre√ßo final com taxa de servi√ßo: R${preco_final:.2f}")
        return preco_final



asa_noturna_Godoy = Lanche_Godoy("Nightwing Burguer", 18.90, ["P√£o australiano", "Hamb√∫rguer duplo", "Cheddar", "Molho picante"])

asa_noturna_Godoy.exibir_dados_Godoy()


preco_atual = asa_noturna_Godoy.get_preco_Godoy()
print(f"üîç Pre√ßo atual: R${preco_atual:.2f}")

asa_noturna_Godoy.set_preco_Godoy(20.00)  # atualiza o pre√ßo
asa_noturna_Godoy.set_preco_Godoy(-10)    # teste com valor inv√°lido


asa_noturna_Godoy.calcular_preco_final_Godoy()
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 4
import sqlite3

class Lanche_Godoy:
    def __init__(self, nome_lanche, preco, ingredientes):
        self.nome_lanche = nome_lanche
        self.__preco = preco
        self.ingredientes = ingredientes

    def exibir_dados_Godoy(self):
        print(f"{self.nome_lanche} - R${self.__preco:.2f}")

    def get_preco_Godoy(self):
        return self.__preco

    def set_preco_Godoy(self, novo_preco):
        if novo_preco > 0:
            self.__preco = novo_preco

class LancheCombo_Godoy(Lanche_Godoy):
    def __init__(self, nome_lanche, preco, ingredientes, bebida):
        super().__init__(nome_lanche, preco, ingredientes)
        self.bebida = bebida

    def exibir_dados_Godoy(self):
        print(f"{self.nome_lanche} (Combo) - R${self.get_preco_Godoy():.2f} + Bebida: {self.bebida}")

conexao = sqlite3.connect("lanchesGodoy.db")
cursor = conexao.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS lanchesGodoy (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    preco REAL,
    bebida TEXT
)
""")

def inserir_lanche_Godoy(nome, preco, bebida):
    cursor.execute("INSERT INTO lanchesGodoy (nome, preco, bebida) VALUES (?, ?, ?)", (nome, preco, bebida))
    conexao.commit()

def listar_lanches_Godoy():
    cursor.execute("SELECT * FROM lanchesGodoy")
    resultados = cursor.fetchall()
    for lanche in resultados:
        print(f"ID: {lanche[0]} | Nome: {lanche[1]} | Pre√ßo: R${lanche[2]:.2f} | Bebida: {lanche[3]}")

def deletar_lanche_Godoy(id_lanche):
    cursor.execute("DELETE FROM lanchesGodoy WHERE id = ?", (id_lanche,))
    conexao.commit()

batman_combo_Godoy = LancheCombo_Godoy("Batman Combo", 25.90, ["P√£o preto", "Hamb√∫rguer artesanal"], "Refrigerante de uva")
batman_combo_Godoy.exibir_dados_Godoy()

inserir_lanche_Godoy(batman_combo_Godoy.nome_lanche, batman_combo_Godoy.get_preco_Godoy(), batman_combo_Godoy.bebida)

listar_lanches_Godoy()

deletar_lanche_Godoy(1)

conexao.close()
```
