## PARTE 2 – Métodos e Herança

**Tarefas:**
- Criar pelo menos 2 métodos de ação para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo método exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**Código:**

```python
# Desenvolva aqui o código da Parte 2

class Lanche_Godoy:
    def __init__(self, nome_lanche, preco, ingredientes):
        self.nome_lanche = nome_lanche
        self.preco = preco
        self.ingredientes = ingredientes

    def exibir_dados_Godoy(self):
        print(f"🍔 Lanche: {self.nome_lanche}")
        print(f"💰 Preço: R${self.preco:.2f}")
        print("📋 Ingredientes:")
        for ingrediente in self.ingredientes:
            print(f" - {ingrediente}")
        print("-" * 30)

    def aplicar_desconto_Godoy(self, porcentagem):
        desconto = self.preco * (porcentagem / 100)
        self.preco -= desconto
        print(f"🔻 Desconto de {porcentagem}% aplicado! Novo preço: R${self.preco:.2f}")

    def adicionar_ingrediente_Godoy(self, novo_ingrediente):
        self.ingredientes.append(novo_ingrediente)
        print(f"➕ Ingrediente '{novo_ingrediente}' adicionado com sucesso!")



class LancheCombo_Godoy(Lanche_Godoy):
    def __init__(self, nome_lanche, preco, ingredientes, bebida):
        super().__init__(nome_lanche, preco, ingredientes)
        self.bebida = bebida  # novo atributo

    def exibir_combo_Godoy(self):
        self.exibir_dados_Godoy()
        print(f"🥤 Bebida inclusa: {self.bebida}")
        print("=" * 30)



batman_combo_Godoy = LancheCombo_Godoy(
    "Batman Combo",
    25.90,
    ["Pão preto", "Hambúrguer artesanal", "Queijo gruyère", "Molho secreto da caverna"],
    "Refrigerante de uva"
)


batman_combo_Godoy.exibir_combo_Godoy()
batman_combo_Godoy.aplicar_desconto_Godoy(10)
batman_combo_Godoy.adicionar_ingrediente_Godoy("Bat-bacon")
batman_combo_Godoy.exibir_combo_Godoy()
```

## PARTE 3 – Encapsulamento e Abstração

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar métodos `get` e `set` para o atributo privado.
- Implementar um método que contenha lógica interna (ex: cálculo, verificação).

**Código:**

```python
# Desenvolva aqui o código da Parte 3

class Lanche_Godoy:
    def __init__(self, nome_lanche, preco, ingredientes):
        self.nome_lanche = nome_lanche
        self.__preco = preco  # atributo privado
        self.ingredientes = ingredientes

    def exibir_dados_Godoy(self):
        print(f"🍔 Lanche: {self.nome_lanche}")
        print(f"💰 Preço: R${self.__preco:.2f}")
        print("📋 Ingredientes:")
        for ingrediente in self.ingredientes:
            print(f" - {ingrediente}")
        print("-" * 30)

  
    def get_preco_Godoy(self):
        return self.__preco

    def set_preco_Godoy(self, novo_preco):
        if novo_preco > 0:
            self.__preco = novo_preco
            print(f"✅ Preço atualizado para R${self.__preco:.2f}")
        else:
            print("❌ Preço inválido! Tem que ser maior que zero.")

  
    def calcular_preco_final_Godoy(self, taxa=2.50):
        preco_final = self.__preco + taxa
        print(f"📦 Preço final com taxa de serviço: R${preco_final:.2f}")
        return preco_final



asa_noturna_Godoy = Lanche_Godoy("Nightwing Burguer", 18.90, ["Pão australiano", "Hambúrguer duplo", "Cheddar", "Molho picante"])

asa_noturna_Godoy.exibir_dados_Godoy()


preco_atual = asa_noturna_Godoy.get_preco_Godoy()
print(f"🔍 Preço atual: R${preco_atual:.2f}")

asa_noturna_Godoy.set_preco_Godoy(20.00)  # atualiza o preço
asa_noturna_Godoy.set_preco_Godoy(-10)    # teste com valor inválido


asa_noturna_Godoy.calcular_preco_final_Godoy()
```

## PARTE 4 – Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um método na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inserção (INSERT), consulta (SELECT) e exclusão (DELETE) de registros no banco de dados.

**Código:**

```python
# Desenvolva aqui o código da Parte 4
import sqlite3

class Lanche_Godoy:
    def __init__(self, nome_lanche, preco, ingredientes):
        self.nome_lanche = nome_lanche
        self.__preco = preco
        self.ingredientes = ingredientes

    def exibir_dados_Godoy(self):
        print(f"{self.nome_lanche} - R${self.__preco:.2f}")

    def get_preco_Godoy(self):
        return self.__preco

    def set_preco_Godoy(self, novo_preco):
        if novo_preco > 0:
            self.__preco = novo_preco

class LancheCombo_Godoy(Lanche_Godoy):
    def __init__(self, nome_lanche, preco, ingredientes, bebida):
        super().__init__(nome_lanche, preco, ingredientes)
        self.bebida = bebida

    def exibir_dados_Godoy(self):
        print(f"{self.nome_lanche} (Combo) - R${self.get_preco_Godoy():.2f} + Bebida: {self.bebida}")

conexao = sqlite3.connect("lanchesGodoy.db")
cursor = conexao.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS lanchesGodoy (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    preco REAL,
    bebida TEXT
)
""")

def inserir_lanche_Godoy(nome, preco, bebida):
    cursor.execute("INSERT INTO lanchesGodoy (nome, preco, bebida) VALUES (?, ?, ?)", (nome, preco, bebida))
    conexao.commit()

def listar_lanches_Godoy():
    cursor.execute("SELECT * FROM lanchesGodoy")
    resultados = cursor.fetchall()
    for lanche in resultados:
        print(f"ID: {lanche[0]} | Nome: {lanche[1]} | Preço: R${lanche[2]:.2f} | Bebida: {lanche[3]}")

def deletar_lanche_Godoy(id_lanche):
    cursor.execute("DELETE FROM lanchesGodoy WHERE id = ?", (id_lanche,))
    conexao.commit()

batman_combo_Godoy = LancheCombo_Godoy("Batman Combo", 25.90, ["Pão preto", "Hambúrguer artesanal"], "Refrigerante de uva")
batman_combo_Godoy.exibir_dados_Godoy()

inserir_lanche_Godoy(batman_combo_Godoy.nome_lanche, batman_combo_Godoy.get_preco_Godoy(), batman_combo_Godoy.bebida)

listar_lanches_Godoy()

deletar_lanche_Godoy(1)

conexao.close()
```

```python
# Desenvolva aqui o código da Parte 5
# 📘 Projeto Final – Sistema de Vendas de Lanches Godoy 🦇

import sqlite3

# ========================
# CLASSE PRINCIPAL
# ========================

class Lanche_Godoy:
    def __init__(self, nome_lanche, preco, ingredientes):
        self.nome_lanche = nome_lanche
        self.__preco = preco  # atributo privado (encapsulamento)
        self.ingredientes = ingredientes

    def exibir_dados_Godoy(self):
        print(f"🍔 Lanche: {self.nome_lanche} - R${self.__preco:.2f}")
        print("📋 Ingredientes:")
        for item in self.ingredientes:
            print(f" - {item}")
        print("-" * 30)

    def get_preco_Godoy(self):
        return self.__preco

    def set_preco_Godoy(self, novo_preco):
        if novo_preco > 0:
            self.__preco = novo_preco
        else:
            print("❌ Preço inválido")

    def calcular_preco_final_Godoy(self, taxa=2.50):
        return self.__preco + taxa


# ========================
# SUBCLASSE COM HERANÇA
# ========================

class LancheCombo_Godoy(Lanche_Godoy):
    def __init__(self, nome_lanche, preco, ingredientes, bebida):
        super().__init__(nome_lanche, preco, ingredientes)
        self.bebida = bebida  # novo atributo

    def exibir_dados_Godoy(self):  # polimorfismo: sobrescrevendo método
        print(f"🍟 Combo: {self.nome_lanche} - R${self.get_preco_Godoy():.2f} + Bebida: {self.bebida}")
        print("📋 Ingredientes:")
        for item in self.ingredientes:
            print(f" - {item}")
        print("-" * 30)


# ========================
# BANCO DE DADOS (SQLite)
# ========================

conexao = sqlite3.connect("lanchesGodoy.db")
cursor = conexao.cursor()

# Criação da tabela
cursor.execute("""
CREATE TABLE IF NOT EXISTS lanchesGodoy (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    preco REAL,
    bebida TEXT
)
""")

# Inserção de lanche
def inserir_lanche_Godoy(nome, preco, bebida):
    cursor.execute("INSERT INTO lanchesGodoy (nome, preco, bebida) VALUES (?, ?, ?)", (nome, preco, bebida))
    conexao.commit()

# Listar lanches
def listar_lanches_Godoy():
    cursor.execute("SELECT * FROM lanchesGodoy")
    resultados = cursor.fetchall()
    print("\n📦 Lanches no Banco de Dados:")
    for lanche in resultados:
        print(f"ID: {lanche[0]} | Nome: {lanche[1]} | Preço: R${lanche[2]:.2f} | Bebida: {lanche[3]}")
    print("-" * 30)

# Deletar lanche
def deletar_lanche_Godoy(id_lanche):
    cursor.execute("DELETE FROM lanchesGodoy WHERE id = ?", (id_lanche,))
    conexao.commit()


# ========================
# TESTES FINAIS
# ========================

# Criando objetos
lanche1 = Lanche_Godoy("Nightwing Burguer", 18.90, ["Pão australiano", "Hambúrguer duplo", "Cheddar"])
lanche2 = LancheCombo_Godoy("Robin Combo", 24.90, ["Pão de brioche", "Frango empanado", "Alface", "Tomate"], "Suco de laranja")

# Exibindo dados
lanche1.exibir_dados_Godoy()
lanche2.exibir_dados_Godoy()

# Usando encapsulamento
lanche1.set_preco_Godoy(20.00)
print(f"🔎 Preço final do {lanche1.nome_lanche}: R${lanche1.calcular_preco_final_Godoy():.2f}")

# Inserindo no banco
inserir_lanche_Godoy(lanche2.nome_lanche, lanche2.get_preco_Godoy(), lanche2.bebida)

# Consultando banco
listar_lanches_Godoy()

# Excluindo lanche (exemplo com ID 1, muda conforme seu BD)
deletar_lanche_Godoy(1)

# Fechando conexão
conexao.close()
```
